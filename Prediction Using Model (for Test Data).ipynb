### Making Prediction of Test Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
app_test = pd.read_csv("home-credit-default-risk/application_test.csv")
app_test.head(2)
```

```python
# Define all the variables that is needed 
# (num_to_cat, correlated_var, list_cat_ok, ohe, dict_standardization, list_num, num_mean, dict_cat_mode)
# load from disk
import joblib

num_to_cat = joblib.load("num_to_cat.pkl")
correlated_var = joblib.load("correlated_var.pkl")
list_cat_ok = joblib.load("list_cat_ok.pkl")
ohe = joblib.load("one_hot_encoder.pkl")
dict_standardization = joblib.load("dict_standardization.pkl")
list_num = joblib.load("list_num.pkl")
num_mean = pd.read_csv("num_mean.csv", index_col = 0)
dict_cat_mode = joblib.load("dict_cat_mode.pkl")
```

```python
# 1. Change dtype
app_test[num_to_cat] = app_test[num_to_cat].astype(object)

# 2. Replace several data
# "XNA" in CODE_GENDER replace by NAN
# "Unknown" in NAME_FAMILY_STATUS replace by NAN
# "XNA" in ORGANIZATION_TYPE replace by "Other"
app_test["CODE_GENDER"].replace("XNA", np.nan, inplace = True)
app_test["NAME_FAMILY_STATUS"].replace("Unknown", np.nan, inplace = True)
app_test["ORGANIZATION_TYPE"].replace("XNA", "Other", inplace = True)

# 3. Impute missing data with value from train data
# impute missing value in numerical data using "mean"
app_test[list_num] = app_test[list_num].fillna(num_mean["0"])

# impute missing value in categorical data using "mode"
for i, m in dict_cat_mode.items():
    app_test[i] = app_test[i].fillna(m)

# 4. Correlated variables
correlated_var2 = correlated_var
correlated_var2.remove("TARGET")
app_test = app_test[correlated_var2]

app_test_ok = app_test.copy()

# there is 'REGION_RATING_CLIENT_W_CITY_-1' that shouldn't be exist (Range of the value just beetwen 1-3)
app_test_ok["REGION_RATING_CLIENT_W_CITY"].replace(-1, 1, inplace = True)
```

```python
# 5. OneHotEncoder transform
# fit OneHotEncoder for train data first
data_train_cor = pd.read_csv("data train with correlated var.csv", index_col = 0)
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse = False, handle_unknown = "ignore")
encoded_nominal = ohe.fit(data_train_cor[list_cat_ok])

# transform OneHotEncoder for test data
encoded_test = ohe.transform(app_test_ok[list_cat_ok])

# converting to a dataframe 
encoded_test_df = pd.DataFrame(encoded_test, columns = ohe.get_feature_names_out())

# combine with the original data
app_test_ok = pd.concat(objs = [encoded_test_df, app_test_ok], axis = 1)
app_test_ok.drop(list_cat_ok, axis = 1, inplace = True)

# 6. Feature scaling
for i, s in dict_standardization.items():
    loaded_stand = joblib.load(s)
    # transform the testing data column
    app_test_ok[i] = loaded_stand.transform(app_test_ok[[i]])
```

```python
# 7. Predict testing data
id_curr_test = app_test_ok["SK_ID_CURR"]
app_test_ok = app_test_ok.drop("SK_ID_CURR", axis = 1)

# load the model
logistic_model = joblib.load("model_logistic_regression_ok.pkl")
forest_model = joblib.load("model_random_forest_ok.pkl")

# make prediction
y_pred_logistic = logistic_model.predict(app_test_ok)
y_pred_forest = forest_model.predict(app_test_ok)

# combine prediction value with the ID
pred_logistic = pd.DataFrame(y_pred_logistic, columns = ["TARGET"])
pred_forest = pd.DataFrame(y_pred_forest, columns = ["TARGET"])
pred_logistic_df = pd.concat(objs = [id_curr_test, pred_logistic], axis = 1)
pred_forest_df = pd.concat(objs = [id_curr_test, pred_forest], axis = 1)

# save the result of prediction
pred_logistic_df.to_csv("result_logistic_prediction.csv", index = False)
pred_forest_df.to_csv("result_forest_prediction.csv", index = False)
```
